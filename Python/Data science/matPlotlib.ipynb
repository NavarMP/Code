```python
name = ['hashir', 'hadi', 'navar', 'asif']
mark = [10, 5, 50, 60]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot (name, mark, marker='.')
plt.xlabel ("Name")
plt.ylabel ("Mark")
plt.show()
```

```python
name = ['hadi', 'hashir', 'NavarMP', 'Asif']
mark = [49, 34, 28, 60]
colors = ['#a4133c', '#ff4d6d','#ff8fa3', '#590d22']
plt.bar (name, mark, color = colors, width = 0.5)
plt.grid(color = '#121212')

plt.show()
```

```python
plt.scatter (name, mark, marker = ',')
```

```python
list = [67, 45, 20, 12]
name = ['Hashir', 'Hadi', 'Navar', 'Asif']
colors = ['#a4133c', '#ff4d6d','#ff8fa3', '#590d22']
plt.pie (list, labels = name, colors=colors)
plt.legend()
```
